```python
import torch
import matplotlib.pyplot as plt
import math

torch.manual_seed(42)

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

print(device)
```

```python
tokens = ["The", "cat", "sat", "on", "mat"]
seq_len = len(tokens)
d_model = 8
```

```python
x = torch.randn(seq_len, d_model, device=device)
print("Tokens:", tokens)
print("Embedding Shape:", x.shape)
```

```python
d_k = 8

W_q = torch.randn(d_model, d_k, device=device)
W_k = torch.randn(d_model, d_k, device=device)
W_v = torch.randn(d_model, d_k, device=device)

print("W_q:", W_q.shape)
print("W_k:", W_k.shape)
print("W_v:", W_v.shape)
```

```python
Q = x @ W_q
K = x @ W_k
V = x @ W_v

print("Q:", Q.shape)
print("K:", K.shape)
print("V:", V.shape)
```

```python
print("Token:", tokens[1])
print("Embedding:", x[1])

print("\nQuery:")
print(Q[1])

print("\nKey:")
print(K[1])

print("\nValue:")
print(V[1])
```
